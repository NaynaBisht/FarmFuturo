```python
import pandas as pd
import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python


data = {
    'Model': ['DecisionTree', 'LogisticRegression', 'RandomForest', 'SVM'],
    'F1Score': [0.9591076, 0.667221902, 0.970032, 0.90078],
    'Accuracy': [0.95909, 0.6772727, 0.9704545, 0.90909],
    'Recall': [0.95909, 0.6772727, 0.9704545, 0.90909],
    'Precision': [0.962854601, 0.677301, 0.9734036, 0.904269],
    'Timetaken': [0.0337, 0.157, 0.7254, 0.0734]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Model'], df['Accuracy'], marker='o', linestyle='-', color='b')
plt.title('Accuracy according to each Model')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Model'], df['F1Score'], marker='o', linestyle='-', color='g')
plt.title('F1Score according to Model')
plt.xlabel('Model')
plt.ylabel('F1Score')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Model'], df['Timetaken'], marker='o', linestyle='-', color='r')
plt.title('Timetaken according to Model')
plt.xlabel('Model')
plt.ylabel('Timetaken')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Model'], df['Precision'], marker='o', linestyle='-', color='purple')
plt.title('Precision according to Model')
plt.xlabel('Model')
plt.ylabel('Precision')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,12))

plt.plot(df['Model'], df['F1Score'], marker='o', linestyle='-', color='b', label='F1Score')
plt.plot(df['Model'], df['Accuracy'], marker='o', linestyle='-', color='g', label='Accuracy')
plt.plot(df['Model'], df['Recall'], marker='o', linestyle='-', color='r', label='Recall')
plt.plot(df['Model'], df['Precision'], marker='o', linestyle='-', color='purple', label='Precision')

plt.xlabel('Model', fontweight='bold')
plt.ylabel('Metric Value')
plt.title('Performance Metrics Comparison according to model')
plt.legend()
plt.grid(True)
plt.show()
```
